```python
import moviepy.editor as mp

filename = "/content/drive/MyDrive/gitlab_engineering_meeting.mp3"
start = 0
end = 600 #24 * 60 + 10

clip = mp.AudioFileClip(filename)
end = min(clip.duration, end)

sub_clip = clip.subclip(start, end)

sub_clip.write_audiofile("audio_short" + ".mp3")

#filename = "audio_short" + ".mp3"
```

```python
# requires high ram runtime
from transformers import WhisperProcessor, WhisperForConditionalGeneration
import librosa


speech, _ = librosa.load(filename)

processor = WhisperProcessor.from_pretrained("openai/whisper-large")
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-large")

model.config.forced_decoder_ids = processor.get_decoder_prompt_ids(language = "en", task = "transcribe")
input_features = processor(speech, return_tensors="pt").input_features 
predicted_ids = model.generate(input_features)
transcription = processor.batch_decode(predicted_ids)

print(processor.batch_decode(predicted_ids, skip_special_tokens = True))
```
